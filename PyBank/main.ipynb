```python
# Import pertinent libraries
from pathlib import Path
import csv


# Establish current directory
current_dir = Path.cwd()
print(current_dir)

# Set path using pathlib
budget_data = current_dir/'02-Homework_02-Python_Instructions_PyBank_Resources_Budget_Data.csv'


# Open csv file as an object
with open(budget_data) as csv_file:
    
    csv_reader = csv.reader(csv_file, delimiter = ',')
    csv_header = next(csv_reader)
    
    # Print header to test
    print(csv_header)
    
    line_count = 0
    
    # Establish variables to be to be utilized while looping through csv rows
    row_count = []
    profit_change = []
    monthly_profits = {}
    total_profit = 0
    previous_month_profit = 0 #Not sure if we should assume 0 for the first month's previous month
    
    # Print the rows of data
    for row in csv_reader:
        print(row)
        
        # Convert row variables for better readability and data types
        date = row[0]
        profit = int(row[1])
        
        # Calculate total number of months included in dataset
        row_count.append(row[0])
        
        
        # Calculate the net total profit/loss of entire period
        total_profit += profit
        
        # Create dictionary of months with their profits and profit changes
        monthly_profits[date] = [profit, profit - previous_month_profit]
        previous_month_profit = profit
        
        #Create list of monthly profit changes
        profit_change.append(monthly_profits[date][1]) 
        
    
    # Calculate average change of profit and loss over entire period
    avg_profit_change = sum(profit_change) / len(profit_change)
    print(avg_profit_change)
        
    # Calculate greatest increase in profits
    max_profit_change = max(profit_change)
    
    #Extract months of highest and lowest profit
    for date in monthly_profits:
        if monthly_profits[date][1] == max(profit_change):
            max_profit_month = date
        elif monthly_profits[date][1] == min(profit_change):
            min_profit_month = date
        
        
    # Calculate greatest decrease in profits
    min_profit_change = min(profit_change)
    
    #Extract month of lowest profit
    
    
    # Print number of rows (months)
    print(len(row_count))
    print(total_profit)
    print(monthly_profits)



# Answers
print(f"Total months: {len(row_count)}")
print(f"Total profit: {total_profit}")
print(f"Average change: {avg_profit_change}")
print(f"Greatest increase in profits: {max_profit_month} ({max_profit_change})")
print(f"Greatest decrease in profits: {min_profit_month} ({min_profit_change})")   

```
